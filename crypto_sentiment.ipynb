## 1. Sentiment Analysis

Use the [newsapi](https://newsapi.org/) to pull the latest news articles for Bitcoin and Ethereum and create a DataFrame of sentiment scores for each coin.

Use descriptive statistics to answer the following questions:
1. Which coin had the highest mean positive score?
2. Which coin had the highest negative score?
3. Which coin had the highest positive score?

```python
# Initial imports
import os
import pandas as pd
from dotenv import load_dotenv
import nltk as nltk
from nltk.corpus import stopwords
nltk.download('vader_lexicon')
from nltk.sentiment.vader import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()
from newsapi import NewsApiClient

%matplotlib inline
```

```python
# Read your api key environment variable
load_dotenv()
key = os.getenv('NEWSAPI')


# YOUR CODE HERE!
```

```python
# Create a newsapi client
news = NewsApiClient(api_key=key)
# YOUR CODE HERE!
```

```python
# Fetch the Bitcoin news articles
articles_btc = news.get_everything(q='Bitcoin', language='en')
# YOUR CODE HERE!
```

```python
# Fetch the Ethereum news articles
articles_eth = news.get_everything(q='Ethereum', language='en')
# articles_eth
# YOUR CODE HERE!
```

```python
# Create the Bitcoin sentiment scores DataFrame
btc_sentiments = []
for article in articles_btc["articles"]:
    try:
        text = article["content"]
        date = article["publishedAt"][:10]
        sentiment = analyzer.polarity_scores(text)
        compound = sentiment["compound"]
        pos = sentiment["pos"]
        neu = sentiment["neu"]
        neg = sentiment["neg"]
        
        btc_sentiments.append({
            "text": text,
            "date": date,
            "compound": compound,
            "positive": pos,
            "negative": neg,
            "neutral": neu
            
        })
        
    except AttributeError:
        pass

btc_df = pd.DataFrame(btc_sentiments)
btc_df
# YOUR CODE HERE!
```

```python
# Create the Ethereum sentiment scores DataFrame
# YOUR CODE HERE!
eth_sentiments = []
for article in articles_eth["articles"]:
    try:
        text = article["content"]
        date = article["publishedAt"][:10]
        sentiment = analyzer.polarity_scores(text)
        compound = sentiment["compound"]
        pos = sentiment["pos"]
        neu = sentiment["neu"]
        neg = sentiment["neg"]
        
        eth_sentiments.append({
            "text": text,
            "date": date,
            "compound": compound,
            "positive": pos,
            "negative": neg,
            "neutral": neu
            
        })
        
    except AttributeError:
        pass
eth_df = pd.DataFrame(eth_sentiments)
eth_df
```

###   Tokenizer

In this section, you will use NLTK and Python to tokenize the text for each coin. Be sure to:
1. Lowercase each word.
2. Remove Punctuation.
3. Remove Stopwords.

```python
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer, PorterStemmer
from string import punctuation
import re
```

```python
# Instantiate the lemmatizer
# YOUR CODE HERE!
lem = WordNetLemmatizer()
# Create a list of stopwords
stop_words = set(stopwords.words('english'))
# YOUR CODE HERE!

# Expand the default stopwords list if necessary
# YOUR CODE HERE!
```

```python
# Complete the tokenizer function
def tokenizer(text):
    """Tokenizes text."""
    
    # Remove the punctuation from text
    print(text)
    res = re.sub(r'[^\w\s]', '', text)

   
    # Create a tokenized list of the words
    words = word_tokenize(text)
    
    
    # Lemmatize words into root words
    words_lem = lem.lemmatize(words)

   
    # Convert the words to lowercase
    words_lem = [i.lower() for i in words_lem]
    
    # Remove the stop words
    tokens = [i for i in words_lem if i not in stop_words]
    
    
    return tokens
```
